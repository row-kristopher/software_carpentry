```python
# correlation of pop dens and povery vis scater plot
data_for_al = health_data.loc[health_data['State'] == 'AL',['Population_Density','Poverty']]
data_for_al.head()
```

```python
lst = [tuple(x) for x in data_for_al.values]
lst
```

```python
#creating scatter plot
scatter_plot = pg.XY(stroke=True)
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
# Add title to the graph
scatter_plot = pg.XY(stroke=True)
scatter_plot.title = 'Population Density vs Poverty Correlation'
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
# adding label to x-axis and y-axis
scatter_plot = pg.XY(stroke=True, x_title='Population Density', y_title='Poverty')
scatter_plot.title = 'Population Density vs Poverty Correlation'
scatter_plot.add('AL',lst)
SVG(scatter_plot.render())
```

```python
# changing the x-axis
scatter_plot = pg.XY(stroke=False, x_title='Population Density', y_title='Poverty') #defines axis's
scatter_plot.title = 'Population Density vs Poverty Correlation' #title
scatter_plot.x_labels = (i for i in range(0,2000,100)) #changes x axis numbering
scatter_plot.add('FL',lstfl) # adds in FL data
scatter_plot.add('AL',lst) # adds in AL data
SVG(scatter_plot.render()) #makes the graph appear
```

```python
#adjusting the x axis range
scatter_plot = pg.XY(stroke=False, x_title='Population Density', y_title='Poverty',\
                     xrange = (0,100)) #defines axis's
scatter_plot.title = 'Population Density vs Poverty Correlation' #title
#scatter_plot.x_labels = (i for i in range(0,2000,100)) #changes x axis numbering
scatter_plot.add('FL',lstfl) # adds in FL data
scatter_plot.add('AL',lst) # adds in AL data
SVG(scatter_plot.render()) #makes the graph appear
```

```python
# adjusting the dot size
scatter_plot = pg.XY(stroke=False, x_title='Population Density', y_title='Poverty',\
                     xrange = (0,100), dots_size=1) #defines axis's
scatter_plot.title = 'Population Density vs Poverty Correlation' #title
#scatter_plot.x_labels = (i for i in range(0,2000,100)) #changes x axis numbering
scatter_plot.add('FL',lstfl) # adds in FL data
scatter_plot.add('AL',lst) # adds in AL data
SVG(scatter_plot.render()) #makes the graph appear
```

```python
# mean of poverty in AL
al = health_data.loc[health_data['State'] == 'AL', ['Poverty']]
poverty_mean_al = al['Poverty'].mean()
poverty_mean_al
```

```python
# generalized function to find mean of any state and any column
def mean_of_column (state_name, column_name):
    state_df = health_data.loc[health_data['State'] == state_name,[column_name]]
    return state_df[column_name].mean()
mean_of_column('TX', 'Poverty')
```

```python
def mean_of_column (state_name, column_name):
    state_df = health_data.loc[health_data['State'] == state_name,[column_name]]
    return state_df[column_name].mean()
mean_of_column('AL', 'Poverty')
```

```python
health_data['State'].unique()
type(health_data['State'].unique())
list(health_data['State'].unique())
```

```python
# vertical bar gragh
bar_chart = pg.Bar()
bar_chart.title = 'Mean of Poverty across all states'
for state in list(health_data['State'].unique()):
    bar_chart.add(state, mean_of_column(state, 'Poverty'))
SVG(bar_chart.render())
```

```python
lst_dc = list(health_data['State'].unique())
lst_dc.index('DC')
del lst_dc[8]
lst_dc
```

```python
bar_chart = pg.Bar()
bar_chart.title = 'Mean of Poverty across all states'
for state in list(lst_dc):
    bar_chart.add(state, mean_of_column(state, 'Population_Density'))
SVG(bar_chart.render())
```

```python
bar_chart = pg.HorizontalBar(legend_at_bottom=True, \
                            legend_at_bottom_columns=11)
bar_chart.title = 'Mean of Poverty across all states'
for state in list(lst_dc):
    bar_chart.add(state, mean_of_column(state, 'Poverty'))
SVG(bar_chart.render())
```
