```python
def celsius_fahrenheit(C):
    # F = (C * 9/5) + 32
    fahrenheit = (C * (9/5)) + 32
    return fahrenheit
```

```python
celsius_fahrenheit()
```

```python
celsius_fahrenheit(100)
```

```python
celsius_fahrenheit(C=25)
```

```python
def fahrenheit_celsius(F):
    # C = (F - 32) * (5 / 9)
    celsius = (F - 32) * (5 / 9)
    return celsius
```

```python
fahrenheit_celsius(77)
```

```python
def fahrenheit_celsius(F = 77): 
    # by putting F = 77, 77 is the default variable
    # when the user adds a value, the default is overwritten
    # C = (F - 32) * (5 / 9)
    celsius = (F - 32) * (5 / 9)
    return celsius
```

```python
fahrenheit_celsius(75)
```

```python
fahr = fahrenheit_celsius()
```

```python
def get_urban(df):
    #df = dataframe (df gets replaced by the file name)
    #using df is a pandas object so i dont need to specify pandas (object oriented programing)
    #variable = df.loc[rows,columns]
    urban = df.loc[df['Population_Density'] >= 1000,
                   :]
    return urban
```
